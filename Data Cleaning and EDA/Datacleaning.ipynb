```python
## Encoding gender values using label encoding
%pip install scikit-learn   
from sklearn.preprocessing import LabelEncoder
lbl_encoder = LabelEncoder()
```

```python
gender_encoded = lbl_encoder.fit_transform(df['gender'])
```

```python
df['gender_encoded'] = gender_encoded
df.drop('gender', inplace=True, axis=1)
df.head()
```

```python
df['city_tier'].value_counts()
```

```python
## Encoding city_tier values using label encoding
df['city_tier_encoded'] = lbl_encoder.fit_transform(df['city_tier'])
df.drop('city_tier', inplace=True, axis=1)
```

```python
## Encoding shopping_preference values using label encoding
df['shopping_preference_encoded'] = lbl_encoder.fit_transform(df['shopping_preference'])
df.drop('shopping_preference', inplace=True, axis=1)
df.head()
```
