```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('/Users/devammpatel/Desktop/Consumer_Shopping_Trend/Dataset/Consumer_Shopping_Trends_2026_updated.csv')
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns       
plt.figure(figsize=(25, 15))
sns.heatmap(df.corr(), annot=True) 
plt.title('Correlation Heatmap')
plt.show()
```

```python
## Visualization of online spend 
df.avg_online_spend.plot(kind='kde', color='blue', figsize=(25, 15))
plt.title('Average Online Spend')
plt.xlabel('Index')
plt.ylabel('Average Online Spend')
plt.show()
```

```python
## Visualization of offline spend 
df.avg_store_spend.plot(kind='kde', color='red', figsize=(25, 15))
plt.title('Average Store Spend')
plt.xlabel('Index')
plt.ylabel('Average Store Spend')
plt.show()
```

```python
# Average online spend vs average store spend
plt.figure(figsize=(25, 15))
sns.scatterplot(x='avg_online_spend', y='avg_store_spend', data=df)
plt.title('Online Spend vs Store Spend')
plt.xlabel('Average Online Spend')
plt.ylabel('Average Store Spend')
plt.show()
```

```python
# Average online spend by age group
plt.figure(figsize=(25, 15))
sns.barplot(x='age', y='avg_online_spend', data=df)
plt.title('Average Online Spend by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Average Online Spend')
plt.show()
```

```python
# Average store spend by age group
plt.figure(figsize=(25, 15))
sns.barplot(x='age', y='avg_store_spend', data=df)
plt.title('Average Store Spend by Age Group')       
plt.xlabel('Age Group')
plt.ylabel('Average Store Spend')
plt.show()
```

```python
# Online spend by monthly income
plt.figure(figsize=(25, 15))
sns.scatterplot(x='monthly_income', y='avg_online_spend', data=df)
plt.title('Online Spend vs Monthly Income')
plt.xlabel('Monthly Income')
plt.ylabel('Average Online Spend')
plt.show()
```

```python
# monthly_store_visit by impulse_buying_score
plt.figure(figsize=(25, 15))
sns.histplot(x='monthly_store_visits', y='impulse_buying_score', data=df)
plt.title('Monthly Store Visit vs Impulse Buying Score')                
plt.xlabel('Monthly Store Visit')
plt.ylabel('Impulse Buying Score')
plt.show()
```

```python
#daily_internet_hours by social_media_hours
plt.figure(figsize=(25, 15))        
sns.scatterplot(x='daily_internet_hours', y='social_media_hours', data=df)
plt.title('Daily Internet Hours vs Social Media Hours')
plt.xlabel('Daily Internet Hours')
plt.ylabel('Social Media Hours')
plt.show()
```

```python
# Now let's remove the outliers and visualize the data again to see if there are any changes in the trends.
# Removing outliers using IQR method
minimum,Q1,median,Q3,maximum = df['avg_online_spend'].quantile([0, 0.25, 0.5, 0.75, 1])
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
lower_bound, upper_bound
#There are no outliers in the avg_online_spend column as all the values are within the lower and upper bounds.
```
